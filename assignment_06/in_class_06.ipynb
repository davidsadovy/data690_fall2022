```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv("/content/wdi_data.csv")
df.head()
```

```python
df_usa = df[df["indicator"] == "SP.POP.TOTL"]
df_usa = df_usa[df_usa["Country Code"] == "USA"]

df_usa.head(20)
```

```python
fig = go.Figure()

trace_usa = go.Bar(
    x=df_usa["Year"],
    y=df_usa["value"]
)

fig.add_trace(trace_usa)

fig.update_layout(
    title="USA Population Growth",
    xaxis={"title":"Year"},
    yaxis={"title":"Population"}
)


fig.show()
```

```python
fig = go.Figure()

trace_usa = go.Scatter(
    x=df_usa["Year"],
    y=df_usa["value"],
    mode="markers+lines"
)

fig.add_trace(trace_usa)

fig.update_layout(
    title="USA Population Growth",
    xaxis={"title":"Year"},
    yaxis={"title":"Population"}
)

fig.show()
```

```python
fig = go.Figure()

trace_0 = go.Line(
    x=df["Year"],
    y=df["value"]
)

fig.add_trace(trace_0)

fig.update_layout(
    title="USA Population Growth",
    xaxis={"title":"Year"},
    yaxis={"title":"Population"}
)

fig.show()
```

```python
df_ind = df[df["indicator"] == "SP.POP.TOTL"]
df_ind = df_ind[df_ind["Country Code"] == "IND"]

df_ind.head(20)
```

```python
fig = go.Figure()

trace_usa = go.Line(
    x=df_usa["Year"],
    y=df_usa["value"],
    name="USA"
)

trace_ind = go.Line(
    x=df_ind["Year"],
    y=df_ind["value"],
    name="India"
)

fig.add_trace(trace_usa)
fig.add_trace(trace_ind)

fig.update_layout(
    title="USA vs India Population Growth",
    xaxis={"title":"Year"},
    yaxis={"title":"Population"}
)

fig.show()
```
