```python
# import required libraries
import pandas as pd
import plotly.express as px
```

```python
# read csv data file from web address into a dataframe
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")

# explore shape of dataframe
df.shape
```

```python
# explore random sample of dataframe
df.sample(5)
```

```python
# select countries to examine
# 2 countries in which I've been stationed: NLD, KOR
# 2 countries that I've visited: AUS, SWE
my_countries = ["NLD", "KOR", "AUS", "SWE"]

# create new dataframe with just my selected countries
df2 = df[df["Country Code"].isin(my_countries)]

# explore shape of dataframe
df2.shape
```

```python
# explore random sample of dataframe
df2.sample(5)
```

```python
# confirm unique country names in dataframe
df2["Country Name"].unique()
```

```python
# explore first 5 lines of dataframe
df2.head()
```

```python
# explore last 5 lines of dataframe
df2.tail()
```

```python
# create new dataframe with just population in my selected countries for 2021 (latest year available)
df3 = df2.query('indicator == "SP.POP.TOTL" and Year == 2021')

df3
```

```python
# compare population in my selected countries in 2021 using bar chart
fig = px.bar(
    df3, 
    x="Country Name", 
    y="value", 
    title="2021 Population of Australia, Republic of South Korea, Netherlands, and Sweden", 
    color="Country Name", 
    template="presentation"
    )

# legend is redundant, so remove it
fig.update_layout(showlegend=False)

fig.show()
```

```python
# compare population in my selected countries in 2021 using pie chart
fig = px.pie(
    df3, 
    names="Country Name", 
    values="value",
    title="2021 Population of Australia, Republic of South Korea, Netherlands, and Sweden", 
    color="Country Name", 
    template="seaborn"
    )

# legend is redundant, so remove it
fig.update_layout(showlegend=False)

fig.update_traces(textposition='inside', textinfo='label+value+percent')

fig.show()
```

```python
# create new dataframe with just population in my selected countries for all avaiable years
df4 = df2.query('indicator == "SP.POP.TOTL"')

# explore first 5 rows of dataframe
df4.head()
```

```python
# compare population of my selected countries for all available years using line chart
fig = px.line(
    df4, 
    x="Year",
    y="value", 
    title="Population of Australia, Republic of South Korea, Netherlands, and Sweden from 1960-2021", 
    color="Country Name", 
    template="ggplot2"
    )

fig.show()
```

```python
# create new dataframe with just the population for all countries in 2021
df7 = df.query('indicator == "SP.POP.TOTL" and Year == 2021')

# explore shape of dataframe
df7.shape
```

```python
# explore the first 5 rows of the dataframe
df7.head()
```

```python
# use groupby method to sum populations of each region
df8 = df7.groupby('Region', as_index=False)['value'].sum()

# explore the dataframe
df8
```

```python
# sort the dataframe by value from highest to lowest
df8 = df8.sort_values(by="value", ascending=False)

# explore the dataframe
df8
```

```python
# rename value column to match world dev example
df8 = df8.rename(columns={'value': '2021 Population, total (group sum)'})

# explore the dataframe
df8
```

```python
# compare population by region in 2021 using bar chart
fig = px.bar(
    df8, 
    x="Region", 
    y="2021 Population, total (group sum)", 
    title="2021 Population by Region", 
    color="Region", 
    template="plotly_dark"
    )

# legend is redundant, so remove it
fig.update_layout(showlegend=False)

fig.show()
```

```python
# compare population by region in 2021 using pie chart
fig = px.pie(
    df8, 
    names="Region", 
    values="2021 Population, total (group sum)",
    title="2021 Population by Region", 
    color="Region", 
    template="plotly_dark"
    )

# legend is redundant, so remove it
fig.update_layout(showlegend=False)

fig.update_traces(textposition='inside', textinfo='label+value+percent')

fig.show()
```

```python
# create new dataframe with just the population for all countries in all years
df9 = df.query('indicator == "SP.POP.TOTL"')

# explore first 5 rows of dataframe
df9.shape
```

```python
# use groupby method to sum populations of each region by year
df9 = df9.groupby(['Region', 'Year'])['value'].sum().reset_index()

# explore random sample of dataframe
df9.sample(5)
```

```python
# compare population by region for all available years using line chart
fig = px.line(
    df9, 
    x="Year", 
    y="2021 Population, total (group sum)", 
    title="Annual Population by Region", 
    color="Region", 
    template="plotly_dark"
    )

fig.show()
```

```python
# explore unique indicator names
df["indicator"].unique()
```

```python
# select indicators for Life Expectancy and GDP Per Capita
my_indicators = ['SP.DYN.LE00.IN', 'NY.GDP.PCAP.PP.CD']

# create new dataframe with just the selected indicators for all countries
df5 = df[df["indicator"].isin(my_indicators)]

# explore shape of dataframe
df5.shape
```

```python
# explore random sample of dataframe
df5.sample(10)
```

```python
# reduce dataframe to keep only columns of interest
df5 = df5[["Year", "value", "indicator", "Country Code", "Country Name"]]

# explore random sample of dataframe
df5.sample(5)
```

```python
# create new dataframe with indicators as separate columns
df6 = pd.pivot_table(df5, values="value", index=["Country Code", "Country Name", "Year"], columns="indicator").reset_index()

# explore first 5 rows of dataframe
df6.head()
```

```python
# rename indicator columns
df6 = df6.rename(columns={'NY.GDP.PCAP.PP.CD': 'GDP Per Capita', 'SP.DYN.LE00.IN': 'Life Expectancy'})

# explore first 5 rows of dataframe
df6.head()
```

```python
# clean dataframe by dropping rows without valid data
df6.dropna(inplace=True)

# explore first 5 rows of dataframe
df6.head()
```

```python
# create scatter plot showing health vs wealth for all countries in 2020
fig = px.scatter(
    df6.query('Year == 2020'), 
    x="GDP Per Capita",
    y="Life Expectancy", 
    title="Wealth vs Health", 
    color="Country Name", 
    template="plotly_dark"
    )

# lendend is redundant, so remove it
fig.update_layout(showlegend=False)

fig.show()
```
