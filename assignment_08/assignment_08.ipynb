## Step 6 - Cleanse the data to make the year, population, cost, cost per person numerical type.
- Code functions to cleanse the columns. One function for one column.
- Use `apply()` method of data frame to apply the functions to the columns for data cleansing.

```python
# function to cleanse 1st column

def cleanse_year(a_string):
    '''Converts a string of numbers that may contain asterisks into an integer'''
    return int(a_string.replace("*", ""))
```

```python
# function to cleanse 2nd column

def cleanse_pop(a_string):
    '''Converts a string of numbers that may contain commas into an integer'''
    return int(a_string.replace(",", ""))
```

```python
# function to cleanse 3rd column

def cleanse_cost(a_string):
    '''Converts a string of numbers that may contain dollar signs, commas, 
       and/or the word "Billion" into an integer'''
    new_string = a_string.replace("$", "").replace(",", "")
    if new_string.endswith("Billion"):
        return int(float(new_string.split(" ")[0]) * 10**9)
    else:
        return int(new_string)
```

```python
# function to cleanse 4th column

def cleanse_acpp(a_string):
    '''Converts a string of numbers that may contain dollar signs 
       and/or the word "cents" into an integer'''
    new_string = a_string.replace("$", "")
    if new_string.endswith("cents"):
        return float(new_string.split(" ")[0])/100
    else:
        return float(new_string)
```

```python
# apply() method used to call functions to cleanse data
# cleansed data added to new data frame

df_cleansed = df
df_cleansed["Census Year"] = df["Census Year"].apply(cleanse_year)
df_cleansed["Total Population"] = df["Total Population"].apply(cleanse_pop)
df_cleansed["Census Cost"] = df["Census Cost"].apply(cleanse_cost)
df_cleansed["Average Cost Per Person"] = df["Average Cost Per Person"].apply(cleanse_acpp)
df_cleansed
```

Now that the data is numeric instead of string, the summary statistics are meaningful!

## Step 8 - Use Plotly to plot the average cost per person over time.

```python
fig = px.bar(
    df_cleansed, 
    x="Census Year", 
    y="Average Cost Per Person", 
    title="Average Cost Per Person of the US Census in Dollars <br><sup> 1790-2010</sup>", 
    color="Census Cost", 
    template="presentation"
    )

fig.show()
```

- Note that for the first 150 years or so of the graph, the cost appears to be nearly zero.  
- Early values are overwhelmed by the more recent values.
- Let's try a logarithmic scale on the vertical axis.

```python
fig = px.bar(
    df_cleansed, 
    x="Census Year", 
    y="Average Cost Per Person", 
    title="Average Cost Per Person of the US Census in Dollars <br><sup> LOGARITHMIC SCALE, 1790-2010</sup>", 
    color="Census Cost", 
    template="presentation"
    )

# converts vertical axis to log scale
fig.update_layout(yaxis_type="log")

fig.show()
```

- The log scale allows us to better see the values of the vertical axis.
- The overall shape of the graph may be misleading.
